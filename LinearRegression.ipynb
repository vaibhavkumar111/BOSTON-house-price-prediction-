```python
#Calculate correlation between every column on the data
corr=df.corr('pearson')

#Take absolute values of correlations in a dataframe using list comprehension
corrs= [abs(corr[attr]['target']) for attr in list(features)]

#Make a list of pairs [(corr,features)]
l=list(zip(corrs,list(features)))

#Sort the list of pairs in reverse/descending order 
#with the correlation value as the key for sorting
l.sort(key=lambda x: x[0], reverse=True)

#Unzip pairs to two lists
# zip(*l) -takes a list that looks like [[a,b,c],[d,e,f],[g,h,i]] and returns [[a,d,g][b,e,h][c,f,i]]
corrs,labels=list(zip((*l)))

#Plot correlation with respect to the target variable as a bar graph
index=np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index, corrs, width=0.5)
plt.xlabel('Attributes')
plt.ylabel('Correlation with the target variable')
plt.xticks(index, labels)
plt.show()
```

```python
X=df['LSTAT'].values
Y=df['target'].values
```

```python
#Before normalisation
print(Y[:5])
```

```python
x_scaler=MinMaxScaler()
X=x_scaler.fit_transform(X.reshape(-1,1))
X=X[:,-1]
y_scaler=MinMaxScaler()
Y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=Y[:,-1]
```

```python
#After normalization
print(Y[:5])
```

```python
#0.2=>20% of data is randomly sampled as testing data
xtrain,xtest,ytrain,ytest= train_test_split(X,Y, test_size=0.2)
```

```python
def update(m,x,c,t,learning_rate):
    grad_m=sum(2*((m*x+c)-t)*x)
    grad_c=sum(2*((m*x+c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
```

```python
def error(m,x,c,t):
    N=x.size
    e=sum(((m*x+c)-t)**2)
    return e*1/(2*N)
```

```python
def gradient_descent(init_m,init_c,x,t,learning_rate,iterations,error_threshold):
    m=init_m
    c=init_c
    error_values=list()
    mc_values=list()
    for i in range(iterations):
        e=error(m,x,c,t)
        if e<error_threshold:
            print('Error less than the threshold. Stopping gradient descent')
            break
        error_values.append(e)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
```

```python
%%time
init_m=0.9
init_c=0
learning_rate=0.001
iterations=250
error_threshold=0.001

m,c,error_values,mc_values=gradient_descent(init_m,init_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```

```python
#plot the regression curve
plt.scatter(xtrain,ytrain,color="b")
plt.plot(xtrain, (m*xtrain+c),color="r")
```

```python
#plot error
plt.plot(np.arange(len(mc_values)),error_values)
plt.ylabel("Error")
plt.xlabel("Iterations")
```

```python
#as nmbr of iteration increases, change in line are less noticble
#Inorder to reduce the processing time for the animation, it is advised to chose small value
mc_values_anim=mc_values[0:250:5]
```

```python
#initialize figure and axis
fig,ax=plt.subplots()

#initialize line. The same object is used for init and update functions
ln, =plt.plot([],[],'ro-',animated=True)

#define init fn
def init():
    plt.scatter(xtrain,ytrain,color='b')
    ax.set_xlim(0,1.0)
    ax.set_ylim(0,1.0)
    return ln,

#define update fn
def update_frame(frame):
    m,c=mc_values_anim[frame]
    #Use line eqn to compute end points
    x1,y1=-0.5,m*-.5+c
    x2,y2=1.5,m*1.5+c
    #Set endpoints of line
    ln.set_data([x1,x2],[y1,y2])
    return ln,

#declare animation object. Set callbacks and number of frames.
anim=FuncAnimation(fig,update_frame,frames=range(len(mc_values_anim)),init_func=init,blit=True)

#generate the video of animation that is shown in Jupyter notebook
HTML(anim.to_html5_video())
```

```python
#Calculate the predictions on the test set as a vectorized operation
predicted = (m* xtest)+c
```

```python
#compute MSE for predicted values on testing set
mean_squared_error(ytest,predicted)
```

```python
#put xtest, ytset and predicted values into single dataframe so that
#we can see predicted value alongside the testing set
p = pd.DataFrame(list(zip(xtest,ytest,predicted)),columns=['x','target_y','predicted_y'])
p.head()
```
